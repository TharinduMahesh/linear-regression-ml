```python
import numpy as np
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression

X = 2*np.random.rand(100,1)
y = 1 + 2*X + np.random.randn(100,1);

model = LinearRegression()
model.fit(X,y)

theta_0_sklearn = model.intercept_
theta_1_sklearn = model.coef_

print(f"Scikit-learn found: θ₀ = {theta_0_sklearn[0]}, θ₁ = {theta_1_sklearn[0][0]}")

plt.scatter(X,y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with Scikit-learn")
plt.show()
```

```python

X_b = np.c_[np.ones((100, 1)), X]  

learning_rate = 0.1  
n_iterations = 1000 
m = 100 

theta = np.random.randn(2, 1) 

print(f"Starting with random theta: θ₀ = {theta[0][0]}, θ₁ = {theta[1][0]}")

cost_history = []

for iteration in range(n_iterations):
    predictions = X_b.dot(theta)

    error = predictions - y
    
    cost = (1/ (2*m)) * np.sum(error**2)
    cost_history.append(cost)

    gradient = (1/m) * X_b.T.dot(error)

    theta = theta - learning_rate * gradient


plt.plot(range(n_iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost J(θ)")
plt.title("Cost Function Decreasing Over Time")
plt.show()
```
